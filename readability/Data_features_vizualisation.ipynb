```python
def FRE(texts):
    return [kk.flesch_RE(text) for text in texts]

def FKG(texts):
    return [kk.flesch_kincaid_grade(text) for text in texts]

def SMOG(texts):
    return [kk.smog_index(text) for text in texts]

def CLI(texts):
    return [kk.coleman_liau_index(text) for text in texts]

def DCH(texts):
    return [kk.dale_chall_score(text) for text in texts]

def GF(texts):
    return [kk.gunning_fog(text) for text in texts]
```

```python
#добавляем простые признаки
def avg_sent(texts):
    return [kk.avg_sentence_length(text) for text in texts]

def avg_word_len(texts):
    return [kk.avg_letter_per_word(text) for text in texts]

def avg_sent_len_char(texts):
    return [kk.avg_letter_per_sent(text) for text in texts]

def avg_syllab_word(texts):
    return [kk.avg_syllab_per_word(text) for text in texts]

def avg_syllab_sent(texts):
    return [kk.avg_syllab_per_sent(text) for text in texts]

def percent_syll(text):
    perc_diff = float(float(kk.diffsyll(text)))/float(len(kk.sentence_splitter(text)))
    return round(perc_diff, 2) 


def percent_syll_text(texts):
    return [percent_syll(text) for text in texts]
```

```python
commons = open('common_lemmas.txt', 'r', encoding = 'utf-8')
commons = list(map(lambda x:x.strip(), commons))
```

```python
rudes = open('rude.txt', 'r', encoding = 'utf-8')
rudes = list(map(lambda x:x.strip(), rudes))
```

```python
import re
import pymorphy2 as pm
from stop_words import get_stop_words
```

```python
morph = pm.MorphAnalyzer()
def words(text):
    nolinks = ' '.join([word for word in text.split() if (not re.findall('https?://|\w\.\w', word))])#удалили ссылки
    clean_line = re.sub('[\W\d_-]+', ' ', nolinks.lower().strip())
    ws = re.split(' +', clean_line)
    return [morph.parse(w)[0].normal_form for w in ws]

texts_clean = texts['text'] #прочитали тексты
texts_list = [' '.join(words(text)) for text in texts_clean] # список списков слов
```

```python
texts_list[:10]
```

```python
cleaned = pd.DataFrame({'col':texts_list})
```

```python
cleaned.columns = ['clean']
```

```python
def common_compare(text): #сраниваем текст со списком общих слов
    text_len = len(text.split())
    common_words = 0
    for word in text.split():
        if word in commons:
            common_words += 1
    return round(common_words/text_len, 2) 
```

```python
def total_common(texts):
    return [common_compare(text) for text in texts]
```

```python
def rude_compare(text): #сраниваем текст со списком общих слов
    text_len = len(text.split())
    rude_words = 0
    for word in text.split():
        if word in rudes:
            rude_words += 1
    return round(rude_words/text_len, 2) 
```

```python
def total_rude(texts):
    return [rude_compare(text) for text in texts]
```

```python
scipop_commons = total_common(cleaned['clean'])
```

```python
scipop_commons = pd.DataFrame(scipop_commons)
```

```python
cleaned['common words'] = scipop_commons
```

```python
talks = open('talks.txt', 'r', encoding = 'utf-8')
talks = list(map(lambda x:x.strip(), talks))
```

```python
def talks_compare(text): #сраниваем текст со списком общих слов
    text_len = len(text.split())
    talk_words = 0
    for word in text.split():
        if word in talks:
            talk_words += 1
    return round(talk_words/text_len, 2) 
```

```python
def total_talks(texts):
    return [talks_compare(text) for text in texts]
```

```python
scipop_talks = total_talks(cleaned['clean'])
```

```python
scipop_talks = pd.DataFrame(scipop_talks)
```

```python
cleaned['talk words'] = scipop_talks
```
