```python
import ksslib.readability_metrics as kk
```

```python
def texts_len_sent(texts):
    return [kk.text_len_sent(text) for text in texts]

def texts_len_words(texts):
    return [kk.text_len_words(text) for text in texts]

def texts_len_chars(texts):
    return [kk.char_count(text) for text in texts]

def texts_len_syllabs(texts):
    return [kk.syllable_count(text) for text in texts]

def diff_words(texts):
    return [kk.diffsyll(text) for text in texts]

def avg_word_len(texts):
    return [kk.avg_letter_per_word(text) for text in texts]

def avg_sent(texts):
    return [kk.avg_sentence_length(text) for text in texts]

def avg_sent_len_char(texts):
    return [kk.avg_letter_per_sent(text) for text in texts]

def avg_syllab_word(texts):
    return [kk.avg_syllab_per_word(text) for text in texts]

def avg_syllab_sent(texts):
    return [kk.avg_syllab_per_sent(text) for text in texts]

def perc_syllab_text(texts):
    return [kk.percent_syll(text) for text in texts]
```

```python
SL = texts_len_sent(classes['text'])
WC = texts_len_words(classes['text'])
TC = texts_len_chars(classes['text'])
SYC = texts_len_syllabs(classes['text'])
DW = diff_words(classes['text'])

AWL = avg_word_len(classes['text'])
ASL = avg_sent(classes['text'])
ASLC = avg_sent_len_char(classes['text'])

ASYPW = avg_syllab_word(classes['text'])
ASYPS = avg_syllab_sent(classes['text'])
ADF = perc_syllab_text(classes['text'])
```

```python
SL_df = pd.DataFrame(SL)
WC_df = pd.DataFrame(WC)
TC_df = pd.DataFrame(TC)
SYC_df = pd.DataFrame(SYC)
DW_df = pd.DataFrame(DW)


AWL_df = pd.DataFrame(AWL)
ASL_df = pd.DataFrame(ASL)
ASLC_df = pd.DataFrame(ASLC)

ASYPW_df = pd.DataFrame(ASYPW)
ASYPS_df = pd.DataFrame(ASYPS)
ADF_df = pd.DataFrame(ADF)
```

```python
classes['sentences'] = SL_df
classes['words'] = WC_df
classes['chars'] = TC_df
classes['syllabs'] = SYC_df
classes['diff words'] = DW_df

classes['avg word lenght'] = AWL_df
classes['avg sentence lenght'] = ASL_df
classes['avg sent length-chars'] = ASLC_df


classes['avg syllabs in word'] = ASYPW_df
classes['avg syllabs in sent'] = ASYPS_df
classes['diff words persent'] = ADF_df
```

```python
import ksslib.stat_metrics as km
```

```python
def FRE(texts):
    return [km.flesch_RE(text) for text in texts]

def FKG(texts):
    return [km.flesch_kincaid_grade(text) for text in texts]

def SMOG(texts):
    return [km.smog_index(text) for text in texts]

def CLI(texts):
    return [km.coleman_liau_index(text) for text in texts]

def DCH(texts):
    return [km.dale_chall_score(text) for text in texts]

def GF(texts):
    return [km.gunning_fog(text) for text in texts]
```

```python
scipop_FRE = FRE(classes['text'])
scipop_FKG = FKG(classes['text'])
scipop_SMOG = SMOG(classes['text'])
scipop_CLI = CLI(classes['text'])
scipop_DCH = DCH(classes['text'])
scipop_GF = GF(classes['text'])
```

```python
scipop_FRE = pd.DataFrame(scipop_FRE)
scipop_FKG = pd.DataFrame(scipop_FKG)
scipop_SMOG = pd.DataFrame(scipop_SMOG)
scipop_CLI = pd.DataFrame(scipop_CLI)
scipop_DCH = pd.DataFrame(scipop_DCH)
scipop_GF = pd.DataFrame(scipop_GF)
```

```python
classes['FRE'] = scipop_FRE
classes['FKG'] = scipop_FKG
classes['SMOG'] = scipop_SMOG
classes['CLI'] = scipop_CLI
classes['DCH'] = scipop_DCH
classes['GF'] = scipop_GF
```

```python
import ksslib.readability_dictionary_compare as kd
```

```python
def total_common(texts):
    return [kd.common_compare(text) for text in texts]
```

```python
def total_rude(texts):
    return [kd.rude_compare(text) for text in texts]   
```

```python
def total_talks(texts):
    return [kd.talks_compare(text) for text in texts]  
```

```python
def measure_all(texts):
    return [kd.measure_find(text) for text in texts]
```

```python
def lemma_texts_df(texts):
    texts_list = [' '.join(kd.morphy_words(text)) for text in texts] 
    cleaned = pd.DataFrame({'col':texts_list})
    return cleaned
```

```python
cleaned = lemma_texts_df(classes['text'])
```

```python
cleaned.columns = ['text']
```

```python
commons = total_common(cleaned['text'])
rudes = total_rude(cleaned['text'])
talks = total_talks(cleaned['text'])
measures = measure_all(cleaned['text'])
```

```python
commons_df = pd.DataFrame(commons)
rudes_df = pd.DataFrame(rudes)
talks_df = pd.DataFrame(talks)
measures_df = pd.DataFrame(measures)
```

```python
classes['common words'] = commons_df
classes['rude words'] = rudes_df
classes['talk words'] = talks_df
classes['measures'] = measures_df
```

```python
def all_nouns(texts):
    return [kt.count_noun(text) for text in texts]

def all_verbs(texts):
    return [kt.count_verb(text) for text in texts]

def all_conjs(texts):
    return [kt.count_conj(text) for text in texts]

def all_preps(texts):
    return [kt.count_prep(text) for text in texts]

def all_adjs(texts):
    return [kt.count_adj(text) for text in texts]
```
