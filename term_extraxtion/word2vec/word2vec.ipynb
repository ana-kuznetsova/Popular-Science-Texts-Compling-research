```python
import sys
import gensim, logging

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
corpus = 'string_corpora.txt'
data = gensim.models.word2vec.LineSentence(corpus)
```

```python
%time
model = gensim.models.Word2Vec(data, size=500, window=10, min_count=2, sg=0)
```

```python
print(len(model.wv.vocab))
```

```python
print([w for w in model.wv.vocab][:50])
```

```python
model.save('my.model')
```

```python
model.init_sims(replace=True)
```

```python
import pandas as pd

freq_terms = pd.read_csv('term_freq_sorted.tsv', sep = '\t')
top20 = list(freq_terms['Term'])[:40]
```

```python
print(top20)
```

```python
pos_terms = [term + '_S' for term in top20]
```

```python
for word in top20:
    # есть ли слово в модели? Может быть, и нет
    if word in model:
        print(word)
        # смотрим на вектор слова (его размерность 300, смотрим на первые 10 чисел)
        print(model[word][:10])
        # выдаем 10 ближайших соседей слова:
        for i in model.most_similar(positive=[word], topn=10):
            # слово + коэффициент косинусной близости
            print(i[0], i[1])
        print('\n')
    else:
        # Увы!
        print(word + ' is not present in the model')
```
